```python
#Definición de variables globales
m = 2**32
a = 1013904223
c = 1664525
x0 = (102264 + 102425 + 104112)//3
```

```python
def gcl(n, a = a, x0 = x0, c = c, m = m): #n = cantidad de iteraciones
	valores = []
	x_act = x0
	for i in range(n):
		x_sig = (a * x_act + c) % m
		valores.append(x_sig)
		x_act = x_sig
	return valores
```

b) Modificar el GCL implementado en el punto a) para que devuelva números al azar entre 0 y 1


```python
def gclB(n, a = a, x0 = x0, c = c, m = m): #n = cantidad de iteraciones
	valores_normalizados = gcl(n,a=a,x0=x0,c=c,m=m)
	for i in range(len(valores_normalizados)):
		valores_normalizados[i] = valores_normalizados[i]/m
	return valores_normalizados
```

```python
#Se realiza un test de frecuencia.
def testFrecuencias(n,m,generador):
	pE = 1/m #prob de salir de cada numero
	fEsp = n*pE #frecuencia esperada de cada uno
	nros = collections.Counter(generador(n,m=m))
	dcuad = 0
	for i in range(m):
		dcuad += ((nros[i]-fEsp)**2)
	dcuad = dcuad/fEsp
	limiteSuperior = chi2.ppf(0.99, df=m-1)
	if (dcuad <= limiteSuperior):
		print("El test acepta la hipotesis nula")
	else:
		print("El test rechaza la hipotesis nula")
```

resultados para distintos tamaños de muestra).

```python
# Tomamos inicialmente una normal estándar, tomando de dominio o soporte una exponencial de parametro lambda 1.
# Se debe encontrar el valor máximo (denominado C). Primero calculamos fX(t)/fY(t).
# En cada iteración, se genera un número random x con distribución Exp(λ = 1), se calcula la probabilidad de aceptar como p = fX(x)/c ∗ fY (x).
# Si se acepta el número generado, entonces con probabilidad 1/2 se le cambia el signo

def generar_normal_estandar(N):
    c = 2 * stats.norm.pdf(1) / stats.expon.pdf(1)
    print("El valor maximo es {}".format(c))
    numbers = np.array([])
    n = 0
    while n < N:
        x = np.random.exponential()
        p = stats.norm.pdf(x) / (c * stats.expon.pdf(x))
        y = random.random()

        if y <= p:
            r1 = random.random()
            if r1 < 0.5:
                numbers = np.append(numbers, x)
            else:
                numbers = np.append(numbers, -x)
        
        n += 1

    return numbers

#Aplico el mu y el sigma a los numeros
def generar_normal(n, mu, sigma):
    numbers = generar_normal_estandar(n)
    return (numbers * sigma) + mu

#Se dibuja el histograma
def histograma(numbers):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.hist(numbers, weights=np.zeros_like(numbers) + 1.0 / numbers.size, bins=20)
    plt.show()


# Implemente un método para generar variables aleatorias con distribución normal con media 15 y desvío 3.
n, mu, sigma = 100000, 15, 3
print("Cantidad de números a generar={}".format(n))
print("Distribución Normal(mu={},sigma={})".format(mu, sigma))
numbers = generar_normal(n, mu, sigma)

# Muestre gráficamente la distribución que siguen los números pseudoaleatorios generados.
histograma(numbers)

#Realice 2 tests de los explicados en la materia para verificar si los números generados siguen la distribución pedida (evalué los
#resultados para distintos tamaños de muestra).

#Realizo test de Kolgomorov-Smirnov
def test_normal_kolmogorov_smirnov(numbers_per_test, mu, sigma, alpha):
    print("Hago el test Kolgomorov Smirnov con una muestra de: ",numbers_per_test)
    
    numbers = generar_normal(numbers_per_test, mu, sigma)

    resultado = stats.kstest(numbers, stats.norm(mu,sigma).cdf)

    if (resultado.pvalue < alpha):
		    print("No siguen la distribucion pedida")
    else:
        print("Los datos siguen la distribucion pedida")
    print(" ")
	
def test_normal_anderson(numbers_per_test, mu, sigma):
    print("Hago el test Anderson-Darling con una muestra de: ",numbers_per_test)
    
    numbers = generar_normal(numbers_per_test, mu, sigma)

    resultado = stats.anderson(numbers)

    for i in range(len(resultado.critical_values)):
        nivel_significancia, umbral = resultado.significance_level[i]/100, resultado.critical_values[i]
        if resultado.statistic < resultado.critical_values[i]:
            print('%.3f: %.3f, Datos normales (No se puede rechazar H0)' % (nivel_significancia, umbral))
        else:
            print('%.3f: %.3f, Datos No normales (Se rechaza H0)' % (nivel_significancia, umbral))


alpha = 0.01
numbers_per_test_1 = 100
numbers_per_test_2 = 1000
numbers_per_test_3 = 10000
numbers_per_test_4 = 100000

print(" ")
print(" ")
test_normal_kolmogorov_smirnov(numbers_per_test_1, mu, sigma, alpha)
test_normal_kolmogorov_smirnov(numbers_per_test_2, mu, sigma, alpha)
test_normal_kolmogorov_smirnov(numbers_per_test_3, mu, sigma, alpha)
test_normal_kolmogorov_smirnov(numbers_per_test_4, mu, sigma, alpha)
print(" ")
print(" ")
test_normal_anderson(numbers_per_test_1,mu,sigma)
test_normal_anderson(numbers_per_test_2,mu,sigma)
test_normal_anderson(numbers_per_test_3,mu,sigma)
test_normal_anderson(numbers_per_test_4,mu,sigma)
```

```python
def generar_espacio_vacio():
	espacio = []
	for i in range(NCELDASXFILA):
		fila = []
		for j in range(NCELDASXFILA):
			fila.append("_")
		espacio.append(fila)
	return espacio
```

```python
def hay_movimiento(tipo):
	if (tipo == "A"):
		return True
	probMovimiento = random.random()
	if (tipo == "B" and probMovimiento <= 0.5):
		return True
	if (tipo == "C" and probMovimiento <= 0.25):
		return True
	return False
```

```python
def generar_espacio_vacio():
	espacio = []
	for i in range(NCELDASXFILA):
		fila = []
		for j in range(NCELDASXFILA):
			fila.append("_")
		espacio.append(fila)
	return espacio
```

```python
def obtener_estado(personas):
	espacio = generar_espacio_vacio()
	for v in personas.values():
		espacio[v[0]][v[1]] = v[2]
	return espacio
```

```python
def imprimir_estado(personas):
	print(obtener_estado(personas))
	print("")
```

```python
def hay_movimiento(tipo):
	if (tipo == "A"):
		return True
	probMovimiento = random.random()
	if (tipo == "B" and probMovimiento <= 0.5):
		return True
	if (tipo == "C" and probMovimiento <= 0.25):
		return True
	return False
```

```python
def verificar_rango(i,j,dist):
	if (i + dist < 0 or i + dist >= NCELDASXFILA):
		return False
	if (j + dist < 0 or j + dist >= NCELDASXFILA):
		return False
	return True
```

```python
def verificar_si_se_infecta(espacio,i,j):
	for dist in range(-3,3):
		if (verificar_rango(i,j,dist) == False):
			continue
		if (espacio[i][j+dist] == "X" or espacio[i+dist][j] == "X"):
			probContagio = random.random()
			if (probContagio <= 0.4):
				return True

	for dist in range(-6,-4):
		if (verificar_rango(i,j,dist) == False):
			continue
		if (espacio[i][j+dist] == "X" or espacio[i+dist][j] == "X"):
			probContagio = random.random()
			if (probContagio <= 0.25):
				return True

	for dist in range(4,6):
		if (verificar_rango(i,j,dist) == False):
			continue
		if (espacio[i][j+dist] == "X" or espacio[i+dist][j] == "X"):
			probContagio = random.random()
			if (probContagio <= 0.25):
				return True
	return False
```

```python
def random_walking(personas, contagiados,n,ALPHA,BETA):
	inmunes = []
	dias_enfermedad = {}
	for i in range(n):
		if (personas[i][2] == "X"):
			dias_enfermedad[i] = 1
		else:
			dias_enfermedad[i] = 0

	lista_contagios_por_t = []
	lista_sanos_por_t = []
	lista_contagios_por_t.append(contagiados)
	lista_sanos_por_t.append(n-contagiados)
	for instante in range(INSTANTES):
		for k,v in personas.items():
			if (hay_movimiento(v[3]) == False):
				continue
			estado = obtener_estado(personas)
			movimiento = random.random()
			if (movimiento <= 0.25): #sentido del movimiento
				if (v[1]+1 == NCELDASXFILA): #que no se salga de rango
					continue
				if estado[v[0]][v[1]+1] == '_':#que no este ocupada
					personas[k] = (v[0],v[1]+1,v[2],v[3])
				
			elif (movimiento <= 0.5): #sentido del movimiento
				if (v[1]-1 < 0): #que no se salga de rango
					continue
				if estado[v[0]][v[1]-1] == '_':#que no este ocupada
					personas[k] = (v[0],v[1]-1,v[2],v[3])

			elif (movimiento <= 0.75): #sentido del movimiento
				if (v[0]+1 == NCELDASXFILA): #que no se salga de rango
					continue
				if estado[v[0]+1][v[1]] == '_':#que no este ocupada
					personas[k] = (v[0]+1,v[1],v[2],v[3])

			else: #sentido del movimiento
				if (v[0]-1 < 0): #que no se salga de rango
					continue
				if estado[v[0]-1][v[1]] == '_':#que no este ocupada
					personas[k] = (v[0]-1,v[1],v[2],v[3])


		for i in range(NCELDASXFILA):
			for j in range(NCELDASXFILA):
				if (estado[i][j] == '_'):
					continue
				p = 0
				for k,v in personas.items():
						if (v[0] == i and v[1] == j):
							p = k
							break
				if (estado[i][j] == "O"):
					if (p in inmunes):
						continue
					if (verificar_si_se_infecta(estado,i,j) == True):
						estado[i][j] = "X"
						personas[p] = (i,j,"X",v[3])
						dias_enfermedad[p] = 1
				else:
					dias_enfermedad[p] += 1
					if (dias_enfermedad[p] >= ALPHA):
						probCurarse = random.random()
						if (probCurarse <= BETA): #CURADO
							estado[i][j] = "O"
							personas[p] = (i,j,"O",v[3])
							dias_enfermedad[p] = 0
							inmunes.append(p)

		contagiados_al_dia = 0
		sanos_al_dia = 0
		for k in dias_enfermedad.keys():
			if (dias_enfermedad[k] == 0):
				sanos_al_dia += 1
			else:
				contagiados_al_dia += 1
		lista_contagios_por_t.append(contagiados_al_dia)
		lista_sanos_por_t.append(sanos_al_dia)
	#	imprimir_estado(personas)	
	return lista_contagios_por_t,lista_sanos_por_t				
```

```python
def estado_inicial(n):	
	#Ubico a las N personas de forma aleatoria. El 2% estan infectados.
	#Se decide el "tipo" de persona
	personas = {} #clave: persona. valor: x,y,infectado o no, tipo de persona
	espacio = generar_espacio_vacio()
	contagiados = 0
	for i in range(n):
		x = random.randrange(0,NCELDASXFILA-1)
		y = random.randrange(0,NCELDASXFILA-1)
		while (espacio[x][y] != "_"):
			x = random.randrange(0,NCELDASXFILA-1)
			y = random.randrange(0,NCELDASXFILA-1)

		tipo = ""
		probTipo = random.random()
		if (probTipo <= 0.70):
			tipo = "A"
		elif (probTipo <= 0.95):
			tipo = "B"
		else:
			tipo = "C"

		probInfectado = random.random()
		if (probInfectado <= 0.02):
			personas[i] = (x,y,"X",tipo)
			espacio[x][y]= "X"
			contagiados += 1
		else:
			personas[i] = (x,y,"O",tipo)
			espacio[x][y] = "O"
	return personas, contagiados
```
